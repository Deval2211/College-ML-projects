```python
df['VisitorType']=df['VisitorType'].map({'Returning_Visitor':2,'New_Visitor':1,'Other':0})
df.head()
```

```python
df['Weekend'] = df['Weekend'].map({True: 1, False: 0})
df['Revenue'] = df['Revenue'].map({True: 1, False: 0})
df.head()
```

```python
X = df.drop('Revenue', axis=1)
y = df['Revenue']
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Generate noise with the same shape as X_train
noise = np.random.rand(*X_train.shape)

# Generate integer labels (0 or 1) for y_noise with the same number of samples as X_train
y_noise = np.random.randint(0, 2, size=X_train.shape[0])

# Concatenate X_train and noise to create noisy training data
X_train_noise = np.concatenate([X_train, noise])

# Concatenate y_train and y_noise
y_train_noise = pd.concat([y_train, pd.Series(y_noise)])
```
