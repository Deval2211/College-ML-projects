```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Income'] = le.fit_transform(df['Income'])
df['Income'].value_counts()
```

```python
features = df.drop("Income",axis =1)
y_train = df["Income"]
```

```python
numeric = features.select_dtypes(include="number")
categorical = features.select_dtypes(include="object")
```

```python
for i, column in enumerate(categorical.columns, 1):
    plt.rcParams["figure.figsize"] = (25,15)
    plt.subplot(4,4,i)
    plt.title(column)
    df[column].value_counts(normalize=True).plot(kind='bar')
```

```python
categorical_encoded = categorical.apply(LabelEncoder().fit_transform)

features = pd.concat([numeric, categorical_encoded], axis=1)

display(features.head())
```

```python
features.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
features = sc.fit_transform(features)
X_train = features
```

```python
test_df.columns = df.columns
display(test_df.head())
```

```python
X_test = test_df.drop("Income", axis=1)
y_test = LabelEncoder().fit_transform(test_df["Income"])
```

```python
X_test_numeric = X_test.select_dtypes(include="number")
X_test_categorical = X_test.select_dtypes(include="object")
```

```python
X_test_categorical_encoded = X_test_categorical.apply(LabelEncoder().fit_transform)

X_test = pd.concat([X_test_numeric, X_test_categorical_encoded], axis=1)
```

```python
X_test = sc.transform(X_test)
```

```python
display(X_test.shape)
display(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
lr = LogisticRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
y_pred = gb.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
