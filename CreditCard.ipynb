```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/creditcard.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# checking if there is a null value
df.isnull().sum()
```

```python
#replacing class = 1 to Fraud and same for not Fraud
df['Class'].replace({0:'Not Fraud', 1:'Fraud'}, inplace=True)
```

```python
#counting total Fraud and not Fraud
df['Class'].value_counts()
```

```python
# Bifercating Transaction according to Hours
df['Hour']=df['Time']//60//60
df['Hour'].value_counts().sort_index()
```

```python
# Fraud per Hour
fraud_per_hour = df[df['Class'] == 'Fraud'].groupby('Hour').size()
fraud_per_hour
```

```python
plt.figure(figsize=(12,6))
fraud_per_hour.plot(kind='bar', color='red')
plt.title('Number of Fraud Transactions per Hour')
plt.xlabel('Hour')
plt.ylabel('Count of Fraud Transactions')
plt.show()
```

```python
#Fraud according to Amount
df_fraud = df[df['Class'] == 'Fraud']
bins = [0, 10, 50, 100, 500, 1000, 5000, 10000, df['Amount'].max()+1]
labels = ['0-10', '10-50', '50-100', '100-500', '500-1000', '1k-5k', '5k-10k', '10k+']
df_fraud['AmountRange'] = pd.cut(df_fraud['Amount'], bins=bins, labels=labels, right=False)
count_by_range = df_fraud['AmountRange'].value_counts().sort_index()
plt.figure(figsize=(10,6))
sns.barplot(x=count_by_range.index, y=count_by_range.values, color='red')
plt.title('Fraud Count by Amount Range')
plt.xlabel('Amount Range')
plt.ylabel('Fraud Count')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score,ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

### Diffreciating target variable form pridicting variable

```python
X = df.drop('Class', axis=1)
Y = df['Class']
```

### Making Weight for Bias Data

```python
# giving weight on the this unbalanced data and spliting the data into training variable and target
class_w={'Fraud':0.90,'Not Fraud':0.10}
X_train, X_test, y_train, y_test = train_test_split(
    X, Y,
    test_size=0.2,       # 80% train, 20% test
    random_state=42,     # reproducibility
    stratify=Y           # preserve class distribution
)
```

### Scaling the Data

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Logistic Regression

```python
# using logistic regression
model = LogisticRegression(class_weight=class_w,random_state=0)
model.fit(X_train_scaled, y_train)
```

```python
# creating a confussion Matrix
y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:, 1]
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred,cmap="Blues")
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,pos_label='Fraud')
recall = recall_score(y_test, y_pred,pos_label='Fraud')
f1 = f1_score(y_test, y_pred,pos_label='Fraud')
print("Accuracy:",accuracy)
print("Precision:",precision)
print("Recall:",recall)
print("F1:",f1)
```

Accuracy of the model is 99.91 %

The Precision tells us that the actual Fraud per total Fraud pridicted by model

Recall checks wheather the actual Fraud per total Fraud actually exists in dataset

F1 is the ((recall x precision)/recall + precision)*2

### Decision Tree

```python
# Train the Decision Tree model
dt_model = DecisionTreeClassifier(class_weight=class_w,random_state=42)
dt_model.fit(X_train_scaled, y_train)

# Predict on test data
y_pred_dt = dt_model.predict(X_test_scaled)

# Label Encoder
le = LabelEncoder()
y_test_numeric = le.fit_transform(y_test)
y_pred_dt_numeric = le.transform(y_pred_dt)

# Evaluate the model
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test_numeric, y_pred_dt_numeric)
recall_dt = recall_score(y_test_numeric, y_pred_dt_numeric)
f1_dt = f1_score(y_test_numeric, y_pred_dt_numeric)
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)

print("Decision Tree Classifier Results:")
print(f"Accuracy: {accuracy_dt:.4f}")
print(f"Precision: {precision_dt:.4f}")
print(f"Recall: {recall_dt:.4f}")
print(f"F1 Score: {f1_dt:.4f}")
print("Confusion Matrix:")
print(conf_matrix_dt)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_dt,cmap="Blues")
```
